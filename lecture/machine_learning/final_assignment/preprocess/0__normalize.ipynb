```python
import soynlp
from tqdm import tqdm
import pandas as pd

tqdm.pandas()
```

```python
data = pd.read_csv("../database/review_0.csv", encoding="utf-8")
data[:5]
```

```python
def normalize(x):
    """
    1) organize repeated emoticons appearing
    2) leave only text
    """

    x = soynlp.normalizer.emoticon_normalize(x, num_repeats=2)
    x = soynlp.normalizer.repeat_normalize(x, num_repeats=2)

    return x
```

```python
data["content"] = data["content"].progress_apply(lambda x: normalize(str(x)))
data[0:5]
```

```python
data.to_csv("../database/review_1.csv", encoding="utf-8", index=False)
```
