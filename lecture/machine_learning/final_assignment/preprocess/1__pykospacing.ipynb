```python
from pykospacing import Spacing
from tqdm import tqdm
import pandas as pd
tqdm.pandas()
```

```python
data = pd.read_csv("../database/review_1.csv", encoding="utf-8")
data[:5]
```

```python
spacing = Spacing()
```

```python
def space(x):
    spaced_x = spacing(x)

    return spaced_x
```

```python
data["content"] = data["content"].progress_apply(lambda x: space(str(x)))
```
