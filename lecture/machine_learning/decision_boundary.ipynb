```python
# Import libararies needed

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sympy import Symbol, solve
```

```python
# Show plots

# Subplot 00
# Set p_C1 = p_C2 = 0.5, Set sigma_1 = sigma_2 = 1
p1 = 0.5
p2 = 0.5
mu1 = -2
mu2 = 2
sigma1 = 1
sigma2 = 1

x_values = np.linspace(-10, 10, 1000)
p_x_C1 = stats.norm(mu1, sigma1).pdf(x_values)
p_x_C2 = stats.norm(mu2, sigma2).pdf(x_values)

x = Symbol('x')
equation = (1 / (2 * pow(sigma1, 2)) - 1 / (2 * pow(sigma2, 2))) * pow(x, 2) -\
           (mu1 / (pow(sigma1, 2)) - mu2 / (pow(sigma2, 2))) * x +\
           (pow(mu1, 2) / (2 *  pow(sigma1, 2)) - pow(mu2,2)/(2*pow(sigma2,2))) -\
            np.log((sigma2*p1)/(sigma1*p2))
solution = solve(equation)
decision_boundary = solution

fig = plt.figure(figsize=(20, 10))
ax0 = fig.add_subplot(2, 2, 1)
ax0.plot(x_values, p_x_C1, color='red', label=r'$P(x|C_1)P(C_1)$')
ax0.plot(x_values, p_x_C2, color='blue', label=r'$P(x|C_2)P(C_2)$')
ax0.axvline(decision_boundary, ymin=0, ymax=1, color='black', linestyle='--', label='Decision boundary')
if len(decision_boundary) == 2:
    ax2.axvline(decision_boundary[1], ymin=0, ymax=1, color='black', linestyle='--')
ax0.legend()
ax0.set_xlabel('x')
ax0.set_ylabel('P(x|C)P(C)')
ax0.set_title(r'$p(C_1) = p(C_2)$ = 0.5, ${\sigma_1}^2 = {\sigma_2}^2$ = 1')


# Subplot 01
# Set p_C1 = 0.2, p_C2 = 0.8, Set sigma_1 = sigma_2 = 1
p1 = 0.2
p2 = 0.8
mu1 = -2
mu2 = 2
sigma1 = 1
sigma2 = 1

x_values = np.linspace(-10, 10, 1000)
p_x_C1 = stats.norm(mu1, sigma1).pdf(x_values) * p1
p_x_C2 = stats.norm(mu2, sigma2).pdf(x_values) * p2

x = Symbol('x')
equation = (1 / (2 * pow(sigma1, 2)) - 1 / (2 * pow(sigma2, 2))) * pow(x, 2) -\
           (mu1 / (pow(sigma1, 2)) - mu2 / (pow(sigma2, 2))) * x +\
           (pow(mu1, 2) / (2 *  pow(sigma1, 2)) - pow(mu2,2)/(2*pow(sigma2,2))) -\
            np.log((sigma2*p1)/(sigma1*p2))
solution = solve(equation)
decision_boundary = solution

ax1 = fig.add_subplot(2, 2, 2)
ax1.plot(x_values, p_x_C1, color='red', label=r'$P(x|C_1)P(C_1)$')
ax1.plot(x_values, p_x_C2, color='blue', label=r'$P(x|C_2)P(C_2)$')
ax1.axvline(decision_boundary, ymin=0, ymax=1, color='black', linestyle='--', label='Decision boundary')
if len(decision_boundary) == 2:
    ax2.axvline(decision_boundary[1], ymin=0, ymax=1, color='black', linestyle='--')
ax1.legend()
ax1.set_xlabel('x')
ax1.set_ylabel('P(x|C)P(C)')
ax1.set_title(r'$p(C_1) = 0.2, p(C_2)$ = 0.8, ${\sigma_1}^2 = {\sigma_2}^2$ = 1')


# Subplot 02
# Set p_C1 = p_C2 = 0.5, Set sigma_1 = 1, sigma_2 = 2
p1 = 1
p2 = 1
mu1 = -2
mu2 = 2
sigma1 = 1
sigma2 = 2

x_values = np.linspace(-10, 10, 1000)
p_x_C1 = stats.norm(mu1, sigma1).pdf(x_values) * p1
p_x_C2 = stats.norm(mu2, sigma2).pdf(x_values) * p2

x = Symbol('x')
equation = (1 / (2 * pow(sigma1, 2)) - 1 / (2 * pow(sigma2, 2))) * pow(x, 2) -\
           (mu1 / (pow(sigma1, 2)) - mu2 / (pow(sigma2, 2))) * x +\
           (pow(mu1, 2) / (2 *  pow(sigma1, 2)) - pow(mu2,2)/(2*pow(sigma2,2))) -\
            np.log((sigma2*p1)/(sigma1*p2))
solution = solve(equation)
decision_boundary = solution

ax2 = fig.add_subplot(2, 2, 3)
ax2.plot(x_values, p_x_C1, color='red', label=r'$P(x|C_1)P(C_1)$')
ax2.plot(x_values, p_x_C2, color='blue', label=r'$P(x|C_2)P(C_2)$')
ax2.axvline(decision_boundary[0], ymin=0, ymax=1, color='black', linestyle='--', label='Decision boundary')
if len(decision_boundary) == 2:
    ax2.axvline(decision_boundary[1], ymin=0, ymax=1, color='black', linestyle='--')
ax2.legend()
ax2.set_xlabel('x')
ax2.set_ylabel('P(x|C)P(C)')
ax2.set_title(r'$p(C_1) = p(C_2)$ = 0.5, ${\sigma_1}^2 = 1, {\sigma_2}^2$ = 4')


# Subplot 03
# Set p_C1 = 0.2, p_C2 = 0.8, Set sigma_1 = 1, sigma_2 = 2
p1 = 0.2
p2 = 0.8
mu1 = -2
mu2 = 2
sigma1 = 1
sigma2 = 2

x_values = np.linspace(-10, 10, 1000)
p_x_C1 = stats.norm(mu1, sigma1).pdf(x_values) * p1
p_x_C2 = stats.norm(mu2, sigma2).pdf(x_values) * p2

x = Symbol('x')
equation = (1 / (2 * pow(sigma1, 2)) - 1 / (2 * pow(sigma2, 2))) * pow(x, 2) -\
           (mu1 / (pow(sigma1, 2)) - mu2 / (pow(sigma2, 2))) * x +\
           (pow(mu1, 2) / (2 *  pow(sigma1, 2)) - pow(mu2,2)/(2*pow(sigma2,2))) -\
            np.log((sigma2*p1)/(sigma1*p2))
solution = solve(equation)
decision_boundary = solution

ax3 = fig.add_subplot(2, 2, 4)
ax3.plot(x_values, p_x_C1, color='red', label=r'$P(x|C_1)P(C_1)$')
ax3.plot(x_values, p_x_C2, color='blue', label=r'$P(x|C_2)P(C_2)$')
ax3.axvline(decision_boundary[0], ymin=0, ymax=1, color='black', linestyle='--', label='Decision boundary')
if len(decision_boundary) == 2:
    ax3.axvline(decision_boundary[1], ymin=0, ymax=1, color='black', linestyle='--')
ax3.legend()
ax3.set_xlabel('x')
ax3.set_ylabel('P(x|C)P(C)')
ax3.set_title(r'$p(C_1) = 0.2, p(C_2)$ = 0.8, ${\sigma_1}^2 = 1, {\sigma_2}^2$ = 4')

plt.show()
```
